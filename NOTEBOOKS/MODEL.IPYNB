```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/Users/darshdave/Documents/Projects/ORDER_FOOD_DELIVERY_REVIEW_PREDICTION/DATASET/onlinefoods.csv')
```

```python
#Unnamed Unnamed: 12 has same value so we don't required it
df = df.drop('Unnamed: 12',axis=1)
```

```python
df
```

```python
le = LabelEncoder()
```

```python
#Female : 0
#Male : 1
df['Gender'] = le.fit_transform(df['Gender'])
df.head(5)
```

```python
df['Marital Status'].groupby(df['Marital Status']).count()
#Married : 0
#Prefer not to say : 1
#Single  :2
df['Marital Status'] = le.fit_transform(df['Marital Status'])
```

```python
df['Occupation'].groupby(df['Occupation']).count()
# Employee : 0
# House wife : 1
# Self Employeed : 2
# Student : 3
df['Occupation'] = le.fit_transform(df['Occupation'])
```

```python
df['Monthly Income'].groupby(df['Monthly Income']).count()
# 10001 to 25000 : 0
# 25001 to 50000 : 1
# Below Rs.10000 : 2
# More than 50000 : 3
# No Income : 4
df['Monthly Income'] = le.fit_transform(df['Monthly Income'])
```

```python
df['Educational Qualifications'].groupby(df['Educational Qualifications']).count()
# Graduate         0
# Ph.D             1
# Post Graduate    2
# School           3
# Uneducated       4
df['Educational Qualifications'] = le.fit_transform(df['Educational Qualifications'])
```

```python
df = df.drop(['latitude','longitude'],axis=1)
```

```python
new_df = pd.get_dummies(df, columns=['Pin code'], prefix='Pin code', drop_first=True)
```

```python
new_df
```

```python
new_df['Output'].groupby(new_df['Output']).count()
```

```python
# No     0
# Yes    q
new_df['Output'] = le.fit_transform(new_df['Output'])
```

```python
new_df['Feedback'].groupby(new_df['Feedback']).count()
```

```python
# Negative     0
# Positive     1
new_df['Feedback'] = le.fit_transform(new_df['Feedback'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train = new_df.drop(['Feedback'], axis=1)
y_train = new_df['Feedback']

X_train, X_val, y_train, y_test = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
models = {
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression()
}
```

```python
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "Confusion Matrix": cm
    }
```

```python
for name, result in results.items():
    print(f"Model: {name}")
    print(f"Accuracy: {result['Accuracy']}")
    print(f"Precision: {result['Precision']}")
    print(f"Recall: {result['Recall']}")
    print(f"F1 Score: {result['F1 Score']}")
    print(f"Confusion Matrix:\n{result['Confusion Matrix']}\n")
```

```python
from tabulate import tabulate

headers = ["Model", "Accuracy", "Precision", "Recall", "F1 Score", "Confusion Matrix"]
table = []

for name, result in results.items():
    table.append([
        name,
        result['Accuracy'],
        result['Precision'],
        result['Recall'],
        result['F1 Score'],
    ])

print(tabulate(table, headers=headers, tablefmt='pretty'))
```

```python
best_model_name = max(results, key=lambda x: results[x]['F1 Score'])
best_model = models[best_model_name]

print(f"Best Model: {best_model_name}")
print(f"F1 Score: {results[best_model_name]['F1 Score']}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_neighbors': [3, 5, 7, 9],  # Example values, adjust as needed
    'metric': ['euclidean', 'manhattan']  # Example values, adjust as needed
}

# Initialize KNN classifier
knn = KNeighborsClassifier()

# Initialize GridSearchCV
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='f1')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters found
print("Best Parameters:", grid_search.best_params_)

# Get the best model
best_knn_model = grid_search.best_estimator_
```

```python
best_knn_model = KNeighborsClassifier(metric='manhattan', n_neighbors=5)

# Retrain the model on the entire training dataset
best_knn_model.fit(X_train, y_train)
```

```python
y_pred_val = best_knn_model.predict(X_val)

# Calculate evaluation metrics
accuracy_val = accuracy_score(y_test, y_pred_val)
precision_val = precision_score(y_test, y_pred_val)
recall_val = recall_score(y_test, y_pred_val)
f1_val = f1_score(y_test, y_pred_val)

# Print the evaluation metrics
print("Evaluation Metrics on Validation Set:")
print(f"Accuracy: {accuracy_val}")
print(f"Precision: {precision_val}")
print(f"Recall: {recall_val}")
print(f"F1 Score: {f1_val}")
```
